```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('spam_none_spam_dataset.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.subject = df.subject.str.lower()
```

```python
df.head(20)
```

```python
df.subject = df.subject.str.replace(r'[=>-_?!:,.\[\]\'*"{}()/]', '', regex=True).replace('  ', ' ')
```

```python
df.head()
```

```python
words = set()
for subject in df.subject.tolist():
    for word in subject.split(' '):
        if word == '': continue
        words.add(word.strip())
```

```python
words = list(words)
```

```python
len(words)
```

```python
words.remove('subject')
```

```python
words_dataframe = pd.DataFrame(columns=['subject'] + words)
```

```python
len(words)
```

```python
words_dataframe.head()
```

```python
words_dataframe['subject'] = df.subject
```

```python
words_dataframe.head()
```

```python
words_dataframe = words_dataframe.fillna(0)
words_dataframe.head()
```

```python
words_dataframe.head()
```

```python
count = 0
for word in words:
    try:
        indices = words_dataframe.subject.str.contains(word)
    except Exception as e:
        print('error')
        print(word)
        print(count)
        print(indices.value_counts())
        count += 1
        continue
    words_dataframe.loc[indices, word] = 1
        
```

```python
words_dataframe.head()
```

```python
for column in words_dataframe.columns[1:].tolist():
    print(words_dataframe[column].value_counts())
```

```python
words_dataframe = words_dataframe.drop('subject', axis=1)
```

```python
df.is_spam.value_counts()
```

```python
df.is_spam = df.is_spam.astype(int)
```

```python
words_dataframe['is_spam'] = df.is_spam
```
